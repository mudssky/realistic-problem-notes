```python
def calculate_screen_area(size,aspect_ratio=16/9):
	# 英寸转换为米，因为面积单位里面我们最熟悉的就是平方米
	size_m= size*0.0245
	# 宽高比，即即tanθ，我们可以求cos 和sin
	angel =math.atan(aspect_ratio)
	width=size_m*math.cos(angel)
	height=size_m*math.sin(angel)
	return width*height


```

```python
# screen_sizes= [24,27,32,55,65,75,85,100]
screen_sizes= range(6,100)
screen_areas= [calculate_screen_area(size) for size in screen_sizes]
plt.plot(screen_sizes,screen_areas)
plt.title('屏幕尺寸与面积关系图')
plt.xlabel('屏幕尺寸 (英寸)')
plt.ylabel('屏幕面积 (平方米)')
plt.grid(True)
# 是一个二次函数
```

```python
normal_screen_sizes= [6,8,12,14,16,24,27,32,55,65,75,85,100]
normal_screen_areas= [calculate_screen_area(size) for size in normal_screen_sizes]
def calculate_area_increase(areas):
	res=[]
	length=len(areas)
	for i in range(0,length):
		if i==0:
			res.append(areas[i])
		else:
			res.append(areas[i]-areas[i-1])
	return res

increase_areas= calculate_area_increase(normal_screen_areas)
increase_percentage= [increase_areas[i]/normal_screen_areas[i]*100 for i in range( len(increase_areas))]
compare55=[normal_screen_areas[i]/normal_screen_areas[8] *100 for i in range(len(normal_screen_areas))]
pd.set_option('display.precision',3 )
pd.DataFrame({'尺寸':normal_screen_sizes,'面积':normal_screen_areas,'增加面积':increase_areas,'增加比例':increase_percentage,'相对于55寸':compare55},)
# pd.pre
```
